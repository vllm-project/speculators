```python
import json
import os
```

```python
def saveReduced(filename, save_dir = None):
    with open(filename, 'r') as file:
        data = json.load(file)
    benchmarks=[]

    for i in range(len(data['benchmarks'])):
        benchmarks.append(data['benchmarks'][i]['metrics'])

    actual_filename = filename.split('/')[-1]
    if save_dir is not None:
        actual_filename = os.path.join(save_dir, actual_filename)
    else:
        actual_filename = filename
    with open(actual_filename[:-5]+'_reduced.json', 'w') as f:
        json.dump(benchmarks, f, indent=4)
```

```python
FOLDER1 = "/proving-grounds/machine/linghao/speculators/throughput/Llama-3.1-8B-Instruct-EAGLE3"
FOLDER2 = "/proving-grounds/machine/linghao/speculators/throughput-reduced-EAGLE3"

# for every file in folder1, check if it starts with "output"

total_files = len(os.listdir(FOLDER1))
processed_files = 0

for fname in os.listdir(FOLDER1):
    if not fname.startswith("output"):
        continue
    else:
        saveReduced(os.path.join(FOLDER1, fname), FOLDER2)
    processed_files += 1
    print(f"Processed {processed_files} out of {total_files} files")
```
